## Loading a CSV into a dataframe

```python
import numpy as np
import pandas as pd

perflog_df = pd.read_csv('mssql_3.5.1_perf_log.csv')
```

```python
perflog_df
```

## Examine datatypes

```python
perflog_df.dtypes
```

## Showing a column as a Series

```python
perflog_df.ELAPSED
```

## Remove rows with nulls

```python
perflog_df_no_nulls = perflog_df.query("ELAPSED != '(null)'")
```

```python
perflog_df_no_nulls.head()
```

## Make a narrower dataframe

```python
perflog_df_narrow = perflog_df_no_nulls[['MONITOR', 'ELAPSED']]
```

```python
perflog_df_narrow
```

## Write modified df to csv file

```python
perflog_df_narrow.to_csv('clean_perf_log.csv')
```

```python
%cat clean_perf_log.csv
```

## Change ELAPSED column type

```python
pd.options.mode.chained_assignment = None  # default='warn'
perflog_df_narrow['ELAPSED'] = perflog_df_narrow['ELAPSED'].astype('int')
```

```python
perflog_df_narrow.dtypes
```

## Plot the dataframe

```python
perflog_df_narrow.plot()
```

## Filter on DCS.collectData and plot

```python
perflog_df_narrow_collect = perflog_df_narrow.query('MONITOR == "DCS.collectData"')
```

```python
perflog_df_narrow_collect
```

```python
perflog_df_narrow_collect.plot()
```
